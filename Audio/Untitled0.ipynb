```python
from pyannote.audio import Pipeline

# Replace 'YOUR_AUTH_TOKEN' with your actual token
pipeline = Pipeline.from_pretrained('pyannote/speaker-diarization', use_auth_token='HD_Token')

# Perform diarization
audio_file = '/content/TOEFL Listening Practice Test.mp3'
diarization = pipeline({'uri': 'filename', 'audio': audio_file})

# Print diarization results
for turn, _, speaker in diarization.itertracks(yield_label=True):
    print(f"Speaker {speaker} from {turn.start:.1f}s to {turn.end:.1f}s")
```
