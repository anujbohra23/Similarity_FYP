```python
import os
import shutil


base_dir = "./transcriptions/"
folders = ["txt", "vtt", "srt", "tsv", "json"]

# Create folders if they don't exist
for folder in folders:
    os.makedirs(os.path.join(base_dir, folder), exist_ok=True)

# Move files to their respective folders
for file in os.listdir(base_dir):
    file_path = os.path.join(base_dir, file)
    if os.path.isfile(file_path):
        if file.endswith(".txt"):
            shutil.move(file_path, os.path.join(base_dir, "txt", file))
        elif file.endswith(".vtt"):
            shutil.move(file_path, os.path.join(base_dir, "vtt", file))
        elif file.endswith(".srt"):
            shutil.move(file_path, os.path.join(base_dir, "srt", file))
        elif file.endswith(".tsv"):
            shutil.move(file_path, os.path.join(base_dir, "tsv", file))
        elif file.endswith(".json"):
            shutil.move(file_path, os.path.join(base_dir, "json", file))

for folder in folders:
    folder_path = os.path.join(base_dir, folder)
    if not os.listdir(folder_path):  # Check if folder is empty
        os.rmdir(folder_path)
```
